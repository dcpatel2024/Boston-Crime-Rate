```python
#Top Ten crimes that happned over the years 2016 and 2017
#Check how many counts per unique value of Group

top_10=df['OFFENSE_DESCRIPTION'].value_counts().sort_values(ascending=False).head(10)
top_10
```

```python
##Importing visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# plot the Top Ten common crimes 
# Set the width and height of the figure
plt.figure(figsize=(13,8))

# Add title
plt.title("Top Ten common crimes")

sns.barplot(x=top_10.index,y=top_10)

#Rotate x-labels, otherwise it's utterly hectic
plt.xticks(rotation=60)

# Add label for vertical axis
plt.ylabel("Number of Crimes");
```

```python
##Crimes in each month 
sns.countplot(x ="MONTH", data = df.reset_index()).set(title='Number Of Crimes Each Month')
```

```python
##getting seasons for visualization

def Seasons(mon):
    if (mon == 12 or mon == 1 or mon == 2):
       return "Winter"
    elif(mon == 3 or mon == 4 or mon == 5):
       return "Spring"
    elif(mon ==6 or mon==7 or mon == 8):
       return "Summer"
    else:
       return "Fall"
```

```python
df['Season'] =df.loc[:,'MONTH'].apply(Seasons).sort_values(ascending=False)
df.head()
```

```python
##visualizing in terms of seasons
season_counts = df.groupby('Season')['OFFENSE_CODE'].count().sort_values(ascending=False).to_frame().reset_index()
ax = sns.barplot(x = 'Season' , y = "OFFENSE_CODE",data = season_counts, palette='YlGnBu')
plt.title('Number Of Crimes Each Season');
```

```python
###Crimes district-wise
order = df['DISTRICT'].value_counts(ascending=False).index
sns.countplot(x='DISTRICT', data=df, order=order);
plt.title('The most dangerous places in Boston');
```

```python
##Creating part of day for crime visualization
def day_night(h): 
    if h>=12:
        return "Evening"
    elif h<=5:
        return "Night"
    else:
        return "Morning"
    
```

```python
df['Day/night'] = df.loc[:,'HOUR'].apply(day_night).head()
df.head()
```

```python
##Visualization 
sns.countplot(x=df['Day/night'],order=['Evening','Night','Morning']);
plt.title('Most of crimes hppened durning a day');
```
